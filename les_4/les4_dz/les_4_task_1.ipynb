```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
label_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

title_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

legend_font = {
    "size": 8,
    "family": "monospace",
}
plt.plot(x, y, label="y(x)", color="green", linestyle="solid", marker='*')

plt.title("Задание 1.1", fontdict=title_font)
plt.xlabel("X", fontdict=label_font)
plt.ylabel("Y", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="#000000")
for i_x, i_y in zip(x[:2], y[:2]):
    plt.text(i_x - 0.2, i_y + 0.4, '({}, {})'.format(i_x, i_y), 
             bbox=dict(facecolor='#ffffff', alpha=0.5, edgecolor='#0f0f0f'))
for i_x, i_y in zip(x[2:], y[2:]):
    plt.text(i_x, i_y - 0.5, '({}, {})'.format(i_x, i_y), 
             bbox=dict(facecolor='#ffffff', alpha=0.5, edgecolor='#0f0f0f'))
plt.grid()
plt.show()
```

```python
label_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

title_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

legend_font = {
    "size": 8,
    "family": "monospace",
}
plt.scatter(x, y, label="y(x)", color="green", marker='o')

plt.title("Задание 1.2", fontdict=title_font)
plt.xlabel("X", fontdict=label_font)
plt.ylabel("Y", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="#000000")
plt.grid()
plt.show()
```
