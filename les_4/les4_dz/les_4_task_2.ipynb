```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
label_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

title_font = {
    "color": "#000000",
    "family": "monospace",
    "fontsize": 16,
    "fontweight": "bold",
}

legend_font = {
    "size": 8,
    "family": "monospace",
}
plt.plot(t, f, label="y(x)", color="green", linestyle="solid", marker='.')

plt.title("График f(t)", fontdict=title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="#000000")

plt.axis([0.5, 9.5, -2.5, 2.5])
plt.grid()
plt.show()
```
